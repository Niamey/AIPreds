**BaseLine**

```python
#price_mean = dubai_apartments['pricePerSqmUSD'].mean()
price_mean = dubai_apartments['priceUSD'].mean()
print('MEAN:', price_mean)
#diff = abs(dubai_apartments['pricePerSqmUSD'] - price_mean) / dubai_apartments['pricePerSqmUSD']
diff = abs(dubai_apartments['priceUSD'] - price_mean) / dubai_apartments['priceUSD']
mape = diff.mean()
print('MAPE:', mape)
```

**LINEAR REGRESSION**

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
lr_model = LinearRegression()
# Train your model here:
#lr_train = dubai_apartments[boosting_features].drop('pricePerSqmUSD', axis=1)
#lr_model.fit(lr_train, dubai_apartments['pricePerSqmUSD'])
lr_train = dubai_apartments[boosting_features].drop('priceUSD', axis=1)
lr_model.fit(lr_train, dubai_apartments['priceUSD'])
pred = lr_model.predict(lr_train)
#diff = abs(dubai_apartments['pricePerSqmUSD'] - pred) / dubai_apartments['pricePerSqmUSD']
diff = abs(dubai_apartments['priceUSD'] - pred) / dubai_apartments['priceUSD']
mape = diff.mean()
print('LR MAPE:', mape)
```

**BOOSTING**

```python
import lightgbm as lgb
from sklearn.model_selection import KFold, StratifiedKFold, GroupKFold, StratifiedGroupKFold

#shuffle
dubai_apartments = dubai_apartments.sample(frac=1.0, random_state=42)

dubai_apartments_features_train = dubai_apartments[boosting_features].drop('priceUSD', axis=1)
dubai_apartments_label_train = dubai_apartments['priceUSD']
#dubai_apartments_features_train = dubai_apartments[boosting_features].drop('pricePerSqmUSD', axis=1)
#dubai_apartments_label_train = dubai_apartments['pricePerSqmUSD']

#skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
kf = KFold(n_splits=5, shuffle=True, random_state=42)
#gkf = GroupKFold(n_splits=5)

"""class CustomGroupKFold:
    def __init__(self, n_splits=5, shuffle=False, random_state=None):
        self.gkf = GroupKFold(n_splits=n_splits)

    def split(self, X, y=None, groups=None):
      return self.gkf.split(X, y, groups=dubai_apartments['region'])

gkf = CustomGroupKFold(n_splits=5)"""

num_rounds = 10000

lgb_train = lgb.Dataset(
    dubai_apartments_features_train,
    label=dubai_apartments_label_train,
    categorical_feature=[
        'region',
        'tower',
        #'floor',
        #'furnishing'
    ],
    free_raw_data=False
  )

lgb_parameters = {
    #default
    "objective": "regression",
    "learning_rate": 0.25,
    "num_threads": 10,
    "metric": 'mape',
    "seed": 42,

    #regularization
    "subsample": 0.8,
    "colsample_bytree": 0.8,
    "subsample_freq": 1
}

lgb_result = lgb.cv(
    lgb_parameters,
    lgb_train,
    num_rounds,
    #folds=skf.split(dubai_apartments_features_train, dubai_apartments['region']),
    #folds=gkf.split(dubai_apartments_features_train, dubai_apartments_label_train, groups=dubai_apartments['region']),
    #folds=gkf,
    folds=kf,
    eval_train_metric=True,
    callbacks=[lgb.early_stopping(50), lgb.log_evaluation(period=50)]
  )
```

```python
lgb_model = lgb.train(params=lgb_parameters, train_set=lgb_train, num_boost_round=1528)
```

```python
#boosting_train_features = [x for x in boosting_features if x != 'pricePerSqmUSD']
boosting_train_features = [x for x in boosting_features if x != 'priceUSD']
boosting_train_features
```

```python
feature_importance = lgb_model.feature_importance()

# Print feature importance
for i, importance in sorted(enumerate(feature_importance), key=lambda x: x[1], reverse=True):
    print(f"Feature '{boosting_train_features[i]}': Importance = {importance}")
```
